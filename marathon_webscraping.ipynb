```python
table_19_1_heading = table_19_1.find('tr')
table_19_1_heading = table_19_1_heading.find_all('th')
print(table_19_1_heading)
```

```python
table_19_1_data = table_19_1.find_all('tr')
table_19_1_data.pop(0)
```

```python
table_data = []
```

```python
headings = []
for th in table_19_1_heading:
    headings.append(th.text.replace('\n','').strip())
for tr in table_19_1_data:
    t_row = {}
    for td, th in zip(tr.find_all('td'), headings):
        t_row[th] = td.text.replace('\n', '').strip()
    table_data.append(t_row)
print(table_data)
```

```python
with open('table_data.csv', 'w') as out_file:
    writer = csv.DictWriter(out_file, headings)
    writer.writeheader()
    for row in table_data:
        if row:
            writer.writerow(row)
```

```python
def get_marathon_addresses(address_list, first_address):
    "finds all the page addresses for a R&R Marathon page, and saves it to a list"
    n = 1
    address_n = first_address + str(n)
    resp = requests.post(address_n)
    soup = BeautifulSoup(resp.text, 'html.parser')
    while soup.findAll('li', {'class': 'disabled PagedList-skipToNext'}) == []:
        address_list.append(address_n)
        n += 1
        address_n = first_address + str(n)
        resp = requests.post(address_n)
        soup = BeautifulSoup(resp.text, 'html.parser')
    address_n = first_address + str(n)
    address_list.append(address_n)
    
    return(address_list)
```

```python
def get_marathon_data(data_list, address_list):
    "uses a list of R&R Marathon urls and creates a list of table data from the urls"
    for address in address_list:
        resp = requests.post(address)
        soup = BeautifulSoup(resp.text, 'html.parser')
        table = soup.find('table', {'class': 'table table-responsive table-bordered'})
        table_heading = table.find('tr')
        table_heading = table_heading.find_all('th')
        table_data = table.find_all('tr')
        table_data.pop(0)
        headings = []
        for th in table_heading:
            headings.append(th.text.replace('\n','').strip())
        for tr in table_data:
            t_row = {}
            for td, th in zip(tr.find_all('td'), headings):
                t_row[th] = td.text.replace('\n', '').strip()
            data_list.append(t_row)
    
    return(data_list)
```

```python
def create_marathon_csv(csv_name, headings_list, data_list):
    "creates a csv file from a list of marathon data"
    with open(csv_name, 'w') as out_file:
        writer = csv.DictWriter(out_file, headings_list)
        writer.writeheader()
        for row in data_list:
            if row:
                writer.writerow(row)
        return;
```
